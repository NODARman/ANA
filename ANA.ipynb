# 2. Autocorrelation

```python
def sig_smoothing(n=15, b=[1.0/15]*15, a=1, x=None):
    """See scipy.signal lfilter.
    
    * Filter data along one-dimension with an IIR or FIR filter.
    
    * b : array_like
        The numerator coefficient vector in a 1-D sequence.

    * a : array_like
        The denominator coefficient vector in a 1-D sequence. If a[0] is not 1, then both a and b are normalized by a[0].

    * x : array_like
        An N-dimensional input array.
    
    """

    b = [1.0 / n] * n # b becomes a list with n elements. Example: n=2, b=[0.5, 0.5]
    
    return lfilter(b, a, x)
```

**2.1. Numpy Autocorrelation**

```python
# Autocorrelation
def autocorr(y):
    y = y - np.mean(y) # Zero-centering: Removes the mean value of your light curve
    result = np.correlate(y, y, mode='full')
    result /= result.max() # Norm
    return result[result.size//2:]

ac = autocorr(np.array(lc.flux.value))
# Autocorrelation Period
xx=np.arange(len(ac))*dt # áƒáƒªáƒ¤ áƒ¨áƒ”áƒ˜áƒ«áƒšáƒ”áƒ‘áƒ áƒ˜áƒ›áƒ˜áƒ¢áƒáƒ› áƒáƒ áƒ˜áƒ¡ áƒ“áƒáƒ‘áƒáƒšáƒ˜ áƒ§áƒáƒ•áƒ”áƒšáƒ—áƒ•áƒ˜áƒ¡ áƒ áƒáƒ› 0áƒ“áƒáƒœ áƒ˜áƒ¬áƒ§áƒ”áƒ‘áƒ áƒ›áƒáƒ¡áƒ˜áƒ•áƒ˜ áƒ“áƒ áƒáƒ áƒ dtáƒ“áƒáƒœ, áƒ áƒáƒ“áƒ’áƒáƒœ áƒ£áƒ›áƒáƒ¦áƒšáƒ”áƒ¡áƒ˜ áƒ“áƒ áƒžáƒ˜áƒ áƒ•áƒ”áƒšáƒ˜ áƒžáƒ˜áƒ™áƒ˜ áƒáƒ áƒ˜áƒ¡ áƒœáƒ˜áƒ¥áƒ£áƒ˜áƒ¡áƒ¢áƒ˜áƒ¡ áƒ¡áƒ˜áƒ®áƒ¨áƒ˜áƒ áƒ”...
# áƒáƒ‘áƒ áƒ£áƒ‘áƒ áƒáƒšáƒáƒ“ áƒ“áƒáƒ£áƒ›áƒáƒ¢áƒ” áƒ“áƒ¢ áƒ áƒáƒ› áƒ’áƒáƒ“áƒáƒ˜áƒ¬áƒ˜áƒáƒ¡...

# This manually exponentially decreases the first peak height
# damping=np.linspace(0,max(xx),len(ac))
# ac = ac-np.exp(-damping/2)

ac = sig_smoothing(n=30, x=ac)

peaks, _ = find_peaks(ac, height=0, distance=1/dt)#0.5/dt)
try:
    Period_ACF_Numpy = xx[peaks][1]
except:
    Period_ACF_Numpy = np.nan

# Period_ACF_Numpy = xx[np.where(ac == max(ac[xx>p1*5]))][0]

# Save the value of period
Periods['ACF_Numpy'] = Period_ACF_Numpy
```

**2.2. Scipy Autocorrelation**

```python
sig = lc.flux.value
mask = ~np.isnan(sig)
sig = sig[mask]

# Mean removal (critical)
sig = sig - np.mean(sig)

# Autocorrelation
corr = signal.correlate(sig, sig, mode='full')
lags = signal.correlation_lags(len(sig), len(sig), mode='full')

# Keep non-negative lags
corr = corr[lags >= 0]
lags = lags[lags >= 0]

# Normalize
corr /= corr.max()
ac = corr
# Ignore zero-lag peak


# min_lag = int(0.1 / dt)
# peaks, _ = find_peaks(corr[min_lag:], height=0.1, distance=0.5/dt)
# Period_ACF_Scipy = lags[min_lag:][peaks][0] * dt

xx = lags * dt
ac = sig_smoothing(n=30, x=ac)

peaks, _ = find_peaks(ac, height=0, distance=1/dt)#0.5/dt)
try:
    Period_ACF_Scipy = xx[peaks][1]
except:
    Period_ACF_Scipy = np.nan

Periods['ACF_Scipy'] = Period_ACF_Scipy

print(Period_ACF_Scipy)

plt.figure(figsize=g_size)
plt.plot(xx, ac, color=defcolor, label=starname)
plt.axvline(Period_ACF_Scipy, color='green', linestyle='-', label=f"P = {Period_ACF_Scipy:.4f} Days")
plt.scatter(xx[peaks], ac[peaks], color="red", s=20, label="Peaks")

plt.suptitle("Autocorrelation Function (ACF) Using Scipy")
plt.xlabel("Lag [Days]")
plt.ylabel("ACF Power")
plt.xlim(p1, p2*8 if max(xx)>p2*8 else max(xx))

plt.legend()
save(s2=" _ 2. ACF Scipy")
plt.show()
```

```python

x = lc.time.value # np.linspace(0, 4, 10000)
N=len(x)
y = lc.flux.value # 3*np.sin(51*x*2*np.pi) + 2.1*np.sin(73*x*2*np.pi) + 1.7*np.sin(111*x*2*np.pi) + 10*np.random.rand(N)

T = dt

# remove NaNs
mask = ~np.isnan(x) & ~np.isnan(y)
x, y = x[mask], y[mask]
N = len(y)

# FFT
yf = fft(y) # compute FFT
xf = fftfreq(N, T) # frequencies
xf, yf = xf[xf > 0], 2.0/N*np.abs(yf[xf > 0]) # positive frequencies & normalize
yf /= yf.max()

# select only periods in desired range
mask_period = (1/xf >= p1) & (1/xf <= p2)
xf, yf = xf[mask_period], yf[mask_period]

ps = 1/xf # Periods as ps

# find dominant period
Period_fft = ps[np.argmax(yf)]

plt.figure(figsize=g_size)

plt.plot(ps, yf, color=defcolor, label=starname)
plt.axvline(Period_fft, color="red", linestyle="--", label=f"True Maximum = {round(Period_fft, 2)} Days")

# Optimal peak finder
Thr = 0.2 # Threshold. Default = 0.2 or 20%
pii = np.where(ps==Period_fft)[0] # period initial index
exclude = np.ones(len(ps), dtype=bool)
p_old=0
tempLabel = "Candidate Period(s)"

while ( np.mean([yf[exclude][pii-1],yf[exclude][pii+1]])/yf[exclude][pii]*Thr < np.abs(yf[exclude][pii-1]-yf[exclude][pii+1]) ):
    Period_fft = ps[exclude][np.argmax(yf[exclude])]
    pii = np.where(ps[exclude]==Period_fft)[0] # period initial index
    exclude[pii]=0

    if len(np.where(exclude==False)[0]) > 10 or p_old==Period_fft:
        print("Maximum Peak Finding Limit Reached. Please Tweak The Threshold!")
        break
    
    p_old=Period_fft

    plt.axvline(Period_fft, color="orange", linestyle="--", alpha=0.5, label=tempLabel)
    tempLabel=None

Periods['FFT'] = Period_fft

print(f"ðŸŸ¢ FFT Period: {Period_fft} Days")

plt.axvline(Period_fft, color="green", linestyle="-", label=f"Optimal Period Peak = {round(Period_fft, 2)} Days")
plt.scatter(ps, yf, s=2, color="darkblue")

plt.suptitle("Fast Fourier Transform (FFT)")
plt.xlabel("Period [days]")
plt.ylabel("Normalized Amplitude")
plt.xlim(p1, p2)

plt.legend()
save(s2=" _ 3. FFT")
plt.show()

```

# 4. Time Frequency LombScargle

```python
# Prepare data for 3D TFL plot
def TFL():
    global freq_const, Lombscargles, LombLen, cfrequency, sgnl_n, step
    Signal = lc.flux.value
    tUnit = lc.time.value

    N = len(Signal)
    res_x = 300 # Max number of signals (same as max resolution of x axis). Default (300)
    sgnl_n = 3000 #N//res_x*10 # Length of a signal to be analyzed each time. Default (N//res_x)
    step = 1000 #sgnl_n//10 # Step size for overlapping segments. Default 10% overlap (sgnl_n//10)
    # print(sgnl_n, step)
    # Lomb-Scargle Loop: Compute Lomb-Scargle periodogram for overlapping segments
  

    # Custom Frequency Lombscargle
    cfrequency = np.linspace(1/p2, 1/p1, 5000)
    
    Lombscargles=[]
    start_index=0
    end_index=sgnl_n
    while (end_index) <= N:
        pow_n = LombScargle(tUnit[start_index:end_index], Signal[start_index:end_index]).power(cfrequency)
        Lombscargles.append(pow_n)
        start_index += step
        end_index += step

    LombLen = len(Lombscargles)
    
TFL()
```

```python
# Check lengths of Lombscargle results for inhomogeneity handling
chk_len = [len(lomb) for lomb in Lombscargles]
max_len = max(chk_len)

for i in range(len(chk_len)):
    if chk_len[i] < max_len:
        print(Lombscargles[i], chk_len[i])
        Lombscargles[i] = np.append(Lombscargles[i], np.nan)

Lombscargles = np.array(Lombscargles)

# Prepare parameters for 3D TFL plot
X = np.arange(1, LombLen+1, 1, dtype=int)
X = np.tile(X,(max_len,1))

Y = np.linspace(min(cfrequency), max(cfrequency), max_len)
Y = 1 / np.tile(Y,(LombLen,1)).T

Z = Lombscargles.T
Z = Z / np.max(Z)
```

```python
# Calculate periods of TFL
TFL_Means = np.mean(Lombscargles.T, axis=1)
TFL_RMSs = np.sqrt(TFL_Means**2)

Period_TFL_Mean = 1/cfrequency[np.where(TFL_Means==max(TFL_Means))][0]
Period_TFL_RMS = 1/cfrequency[np.where(TFL_RMSs==max(TFL_RMSs))][0]

Periods["Time_Freq_LS_Mean"] = Period_TFL_Mean
Periods["Time_Freq_LS_RMS"] = Period_TFL_RMS
```

```python
# Autocorrelation
def autocorr(y):
    y = y - np.mean(y) # Zero-centering: Removes the mean value of your light curve
    result = np.correlate(y, y, mode='full')
    result /= result.max() # Norm
    return result[result.size//2:]

# ac = autocorr(np.array(lc.flux.value))

# Autocorrelation Period
# Time_acf = np.arange(len(ac))*dt
# Amplitude_acf = ac / max(ac)

# README: sgnl_n and step variables are DEFINED in the Time Frequency Lombscargle cell
ACFs=[]
si=0 # start_index
# sgnl_n=10000
ei=sgnl_n # end_index
step=step # DEFINED in the Time Frequency Lombscargle cell
# This manually exponentially decreases the first peak height
xx = np.arange(sgnl_n)*dt
while (ei) <= N:
        acf_n = autocorr(np.array(lc.flux.value)[si:ei])
        damping = np.linspace(0,max(xx),len(acf_n))
        acf_n = acf_n-np.exp(-damping/1.5) # This manually exponentially decreases the first peak height
        ACFs.append(acf_n)
        si += step
        ei += step
ACF_len = len(ACFs)

# Check lengths of Lombscargle results for inhomogeneity handling
chk_len = [len(acff) for acff in ACFs]
max_len = max(chk_len)

for i in range(len(chk_len)):
    if chk_len[i] < max_len:
        print(ACFs[i], chk_len[i])
        ACFs[i] = np.append(ACFs[i], np.nan)

ACFs = np.array(ACFs)
ACFs += np.min(abs(ACFs)) # Shift the amplitude to normalize
ACFs /= np.max(ACFs)

# Prepare parameters for 3D TFL plot
XX = np.arange(1, ACF_len+1, 1, dtype=int)
XX = np.tile(XX,(max_len,1))

YY = np.arange(0, len(ACFs[0]))*dt
YY = np.tile(YY,(ACF_len,1)).T

ZZ = ACFs.T
ZZ = np.where(ZZ<0,0,ZZ) # Clamp negative values
# Z += abs(np.min(Z))
# Z = Z / np.max(Z)
```
