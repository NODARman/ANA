```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle
%matplotlib inline
```

# Sygnal

```python
# Lombscargle
a=10
t = np.linspace(0, 2, 1000)
y = np.sin(a*2 * np.pi * t)
```

# 1. Lombscargle

```python
frequency, power = LombScargle(t, y).autopower()

plt.plot(t, y)
plt.show()
plt.plot(frequency, power)
plt.show()

```

# 2. Autocorrelation

```python
# Autocorrelation
def autocorr(y):
    result = np.correlate(y, y, mode='full')
    return result[result.size//2:]
autocorr(y)
plt.plot(autocorr(y))
plt.show()
```

# -1. Periods

```python
print("LombScargle:", max(power))
print("Autocorrelation:", max(y))

print("Delta:", max(power)-max(y))
```
